#2 Evaluation Criteria
The criterion of Mean Absolute Error to assess forecasting performance has both pros and cons. 
Pros:
Compared to mean squared error, the model is less sensitive to outliers, with an error of X simply contributing X to the mean squared error. If we were using mean squared error as our metric, an error of X would be amplified to X^2. This causing a large error to have a significant effect on the models reported error. Meaning the model would likely be too sensitive to outliers An example being an error of 100 being squared to 10000.

Cons:
Not utilizing a percentage based metric causes the severity of errors to be understated specifically on cheaper properties. This as MAE ignores the size of the principle, and is not a relative calculation unlike MAPE which would recognise the severity of an error more effectively. This as an error of 300 should be considered be more significant under a property listed for 500 dollars compared to 2500, due to the size of the error relative to the pricing (60% vs 12%). This meaning the metric may not be the most appropriate due to the commonly wide range of real estate rental prices.

**Task 2, Question 4**: Encode all categorical variables appropriately as discussed in class. 

- Where multiple values are given for an observation encode the observation as 'other'. 
- Where a categorical feature contains more than 5 unique values, map the features into 5 most frequent values + 'other' and then encode appropriately.  
(2 marks)

```python
## Task 2, Question 4 Code Here
#Finding 5 most freqent categories in training columns
import numpy as np
top5={}
for col in cat_columns:
      if train[col].nunique() > 5:
          top5[col] = train[col].value_counts().nlargest(5).index

for col in cat_columns:
    train[col] = train[col].astype(str)
    test[col]  = test[col].astype(str)

for col in cat_columns:
    # mark multi-valued cells as 'other'
    train.loc[train[col].str.contains(',', na=False), col] = 'other'
    test.loc[test[col].str.contains(',',  na=False), col] = 'other'

#Encoding Variables with more than 5 unique variables
for col in top5:
    train[col] = train[col].where(train[col].isin(top5[col]), 'Other')
    test[col] = test[col].where(test[col].isin(top5[col]), 'Other')

    
#Encoding Ordinal Variable (Host_response_time)
response_time_map = { 'within an hour': 1, 'within a few hours': 2, 'within a day': 3, 'a few days or more': 4}
train['host_response_time'] = train['host_response_time'].map(response_time_map)
test['host_response_time'] = test['host_response_time'].map(response_time_map)

#Excluding host_response_time (ordinal variable) from cat_columns before one-hot encoding
cat_columns = list(cat_columns)
cat_columns = [col for col in cat_columns if col != 'host_response_time']


#One hot encoding nominal categorical variables
train = pd.get_dummies(train, columns=cat_columns, drop_first=True)
test = pd.get_dummies(test, columns=cat_columns, drop_first=True)
```

```python
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, RidgeCV, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, make_scorer

#  Split the encoded train into X which are the features and y which is the target price

X = train.drop(columns=['price'])
y = train['price']

#quick check to see minimum and maximum of y range
print("Minimum y:", y.min(), "  Maximum y:", y.max())

#  Set up a 5 fold crossvalidation and an MAE scorer
kf = KFold(n_splits=5, shuffle=True, random_state=42)
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)


# Simple linear regression

linreg = LinearRegression()

# fitting it once on the full train set to inspect coefficients
linreg.fit(X, y)

# extracted intercept and a few example coefficients
print("Simple LR intercept: ", round(linreg.intercept_, 2))

# combining feature names with their learned coefficients
lin_coefs = pd.Series(linreg.coef_, index=X.columns).sort_values(key=abs, ascending=False)

print("\nTop 10 LinearRegression coefficients (by absolute value):")
print(lin_coefs.head(10).round(3))


#test to see what columns is in my X; there should be no price and id
print (X.columns.tolist())

#  calucalte 5 fold CV MAE as a positive number (note : 5 fold CV MAE is just the mean absolute error computed over five cross-validation splits.it is same formual as MAE)
lin_cv_scores = cross_val_score(linreg, X, y, cv=kf, scoring=mae_scorer, n_jobs=-1)
lin_mae  = -lin_cv_scores.mean()

print(f"\nLinearRegression 5fold CV MAE: {lin_mae:.2f}")


# ridge regression 

#  used ridgeCV to try a grid of alpha values and pick the best by 5 fold Cros valifation
alphas = [0.01, 0.1, 1.0, 10.0, 100.0]

ridge_cv = RidgeCV(alphas=alphas, cv=kf, scoring=mae_scorer)
ridge_cv.fit(X, y)

best_alpha = ridge_cv.alpha_
print("\nRidge best alpha (by CV):", best_alpha)

# refit ridge  on the full training data using best alpha
ridge = Ridge(alpha=best_alpha)
ridge.fit(X, y)

ridge_coefs = pd.Series(ridge.coef_, index=X.columns).sort_values(key=abs, ascending=False)
print("\nTop 10 Ridge coefficients (by absolute value):")
print(ridge_coefs.head(10).round(3))

#  Compute 5 fold CV MAE for the selected Ridge model
ridge_cv_scores = cross_val_score(
    Ridge(alpha=best_alpha),
    X, y,
    cv=kf,
    scoring=mae_scorer,
    n_jobs=-1
)
ridge_mae = -ridge_cv_scores.mean()
print(f"\nRidge (={best_alpha}) 5fold CV MAE: {ridge_mae:.2f}")



# 5) Random Forest Regression

rf = RandomForestRegressor(random_state=42)

param_grid = {
    'n_estimators':    [100, 200],
    'max_depth':       [10, 15],
    'min_samples_leaf':[1, 3]
}

grid_rf = GridSearchCV(
    rf,
    param_grid=param_grid,
    scoring=mae_scorer,
    cv=kf,
    n_jobs=-1,
    verbose=1
)
grid_rf.fit(X, y)

best_rf_params = grid_rf.best_params_
print("\nRandom Forest best params:", best_rf_params)

best_rf = grid_rf.best_estimator_

#  feature importances for the Random Forest
rf_importances = (
    pd.Series(best_rf.feature_importances_, index=X.columns)
      .sort_values(ascending=False)
)
print("\nTop 10 Random Forest feature importances:")
print(rf_importances.head(10).round(4))

#  calculates 5 fold cross validatoin MAE of the best RandomForest
rf_mae = -grid_rf.best_score_
print(f"\nRandom Forest bestCV 5fold MAE: {rf_mae:.2f}")


#just some random checks to prepare for my regression model predciton

print(train.dtypes[train.dtypes == "object"])   # make sure no object/texts remain
print(train.columns)                            # double check if dropped price from X
print((train["price"].describe()))              # scale of the target
```

```python
#task 3 question 4 ensemble method way (use this one)
#installing package for xgboost because i dont have it insatlled already
!pip install xgboost --quiet



from sklearn.model_selection import KFold, cross_val_score, GridSearchCV, cross_val_predict
from sklearn.linear_model import Ridge, ElasticNetCV, ElasticNet, LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from sklearn.metrics import mean_absolute_error, make_scorer

#from lightgbm import LGBMRegressor
from xgboost import XGBRegressor

# SPLIT OFF X_full, y_full from TRAIN (7 000 rows), Confirm No object Columns

X_full = train.drop(columns=["price"])
y_full = train["price"].copy()


print(" X_full shape:", X_full.shape)            # expected (7000, n_features)
print(" Any missing in X_full? ", X_full.isna().sum().sum())  
print(" Any object columns in X_full? ", any(X_full.dtypes == "object"))

print(" y_full.min():", y_full.min(), "   y_full.max():", y_full.max())


#  DEFINE 5FOLD MAE SCORER

kf = KFold(n_splits=5, shuffle=True, random_state=42)
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)


#  INSTANTIATE & FIT (Ridge & RF) USING YOUR KNOWN BEST HYPERPARAMETERS
#     no CV here, we assume =100 for Ridge and (200, 15, 3) for RF are final.

#  Ridge with =100
ridge_best = Ridge(alpha=100, random_state=42)
ridge_best.fit(X_full, y_full)
ridge_preds = cross_val_predict(ridge_best, X_full, y_full, cv=kf, n_jobs=-1)
#Inverse Logging
ridge_mae = mean_absolute_error(np.expm1(y_full), np.expm1(ridge_preds))
print(f" Ridge 5-fold CV MAE: {ridge_mae:.2f}")


#  RandomForest with n_estimators=200, max_depth=15, min_samples_leaf=3
rf_best = RandomForestRegressor(
    n_estimators=200,
    max_depth=15,
    min_samples_leaf=3,
    random_state=42
)
rf_best.fit(X_full, y_full)
rf_preds = cross_val_predict(rf_best, X_full, y_full, cv=kf, n_jobs=-1)
#Inverse logging
rf_mae = mean_absolute_error(np.expm1(y_full), np.expm1(rf_preds))
print(f" RandomForest 5-fold CV MAE: {rf_mae:.2f}")

#  ElasticNet (L1+L2) with ElasticNetCV 
# Well try a small grid of l1_ratio and alphas here:
enet_l1_ratios = [0.2, 0.5, 0.8]
enet_alphas    = [0.01, 0.1, 1.0, 10.0]

enet_cv = ElasticNetCV(
    l1_ratio=enet_l1_ratios,
    alphas=enet_alphas,
    cv=kf,
    random_state=42,
    n_jobs=-1
)
enet_cv.fit(X_full, y_full)

best_alpha_enet   = enet_cv.alpha_
best_l1_ratio_enet= enet_cv.l1_ratio_
print(f"\n ElasticNet best : {best_alpha_enet}  best l1_ratio: {best_l1_ratio_enet}")

enet_best = ElasticNet(
    alpha=best_alpha_enet,
    l1_ratio=best_l1_ratio_enet,
    random_state=42,
    max_iter=5000
)
enet_best.fit(X_full, y_full)

# 5-fold CV MAE for ElasticNet:
# Note: ElasticNetCV used internal CV; for a fair comparison we do our own cross_val_score here:
enet_preds = cross_val_predict(enet_best, X_full, y_full, cv=kf, n_jobs=-1)
#Inverse Logging
enet_mae = mean_absolute_error(np.expm1(y_full), np.expm1(enet_preds))
print(f" ElasticNet 5-fold CV MAE: {enet_mae:.2f}")

# Top 5 ENET coefficients by absolute value:
enet_coefs = pd.Series(enet_best.coef_, index=X_full.columns).abs().sort_values(ascending=False)
print("   Top 5 |ElasticNet coefs|:", enet_coefs.head(5).round(2).to_dict())


# === 4.4) Gradient Boosting Regression with GridSearchCV ===
gbr = GradientBoostingRegressor(random_state=42)

param_grid_gbr = {
    "n_estimators":    [100, 200],
    "learning_rate":   [0.05, 0.1],
    "max_depth":       [3, 5],
}
grid_gbr = GridSearchCV(
    gbr,
    param_grid=param_grid_gbr,
    cv=kf,
    scoring=mae_scorer,
    n_jobs=-1,
    verbose=0
)
grid_gbr.fit(X_full, y_full)

best_gbr_params = grid_gbr.best_params_
print(f"\n GBR best params (by CV): {best_gbr_params}")

gbr_best = grid_gbr.best_estimator_

# 5-fold CV MAE for GBR:
gbr_mae = -grid_gbr.best_score_
print(f" GBR bestCV MAE: {gbr_mae:.2f}")

# Top 5 GBR feature importances:
gbr_importances = pd.Series(gbr_best.feature_importances_, index=X_full.columns).sort_values(ascending=False)
print("   Top 5 |GBR importances|:", gbr_importances.head(5).round(3).to_dict())




#  Instantiate a simple XGBRegressor
xgb = XGBRegressor(
    random_state=42,
    objective="reg:squarederror",  # use squared error (standard regression objective)
    verbosity=0
)

#  Define a small grid to tune:
param_grid_xgb = {
    "n_estimators":    [100, 200],
    "learning_rate":   [0.05, 0.1],
    "max_depth":       [3, 5]
}

grid_xgb = GridSearchCV(
    xgb,
    param_grid=param_grid_xgb,
    cv=kf,
    scoring=mae_scorer,
    n_jobs=-1,
    verbose=0
)
grid_xgb.fit(X_full, y_full)

best_xgb_params = grid_xgb.best_params_
print(f"\n XGBoost best params (by CV): {best_xgb_params}")

xgb_best = grid_xgb.best_estimator_
xgb_preds = cross_val_predict(xgb_best, X_full, y_full, cv=kf, n_jobs=-1)
#Inverse logging (as denoted by np.expm1)
xgb_mae = mean_absolute_error(np.expm1(y_full), np.expm1(xgb_preds))
print(f" XGBoost 5-fold CV MAE: {xgb_mae:.2f}")


# -
#  BUILD & EVALUATE THE STACKED MODEL

# Now that we have four tuned base learners:
#     ridge_best    (Ridge with alpha=best_alpha_ridge)
#     rf_best       (RF with best_rf_params)
#     enet_best     (ElasticNet with best alpha & l1_ratio)
#     gbr_best      (GradientBoosting with best_gbr_params)
#
# We will stack them using LinearRegression as the final estimator.

stack_estimators = [
    ("ridge", ridge_best),
    ("rf",    rf_best),
    #("enet",  enet_best),
    ("gbr",   gbr_best),
    # ("lgbm",   lgbm_best),
    ("xgb",     xgb_best), 
]

stack = StackingRegressor(
    estimators       = stack_estimators,
    final_estimator  = LinearRegression(),
    cv               = KFold(n_splits=5, shuffle=True, random_state=42),
    n_jobs           = -1,
    passthrough      = False   # if True, metalearner sees original X as well as base preds
)

# 5-fold CV MAE for the STACK:
stack_preds = cross_val_predict(stack, X_full, y_full, cv=kf, n_jobs=-1)
#Inverse Logging
stack_mae = mean_absolute_error(np.expm1(y_full), np.expm1(stack_preds))
print(f"\n STACK (Ridge+RF+ENet+GBR+XGB) 5-fold CV MAE: {stack_mae:.2f}")


#  RE-FIT ALL MODELS ON THE ENTIRE TRAINING SET

ridge_best.fit(X_full, y_full)
rf_best.fit(X_full, y_full)
#enet_best.fit(X_full, y_full)
gbr_best.fit(X_full, y_full)
#lgbm_best.fit(X_full, y_full)
xgb_best.fit(X_full, y_full)



# Re-fit the stacked ensemble on all 7 000 rows
stack.fit(X_full, y_full)


#  PREDICT ON TEST SET & SAVE submission_ensemble.csv

# The test DataFrame `test` should still contain ID plus exactly all dummy-encoded columns.
# We need to drop ID before predicting so that columns align exactly with X_full.columns.

X_test = test.drop(columns=["ID"], errors="ignore")

#  VERY CRITICAL: reindex X_test to have exactly the same columns & order as X_full 
X_test = X_test.reindex(columns=X_full.columns, fill_value=0)

assert list(X_test.columns) == list(X_full.columns), (" Columns must match EXACTLY before stack.predict!")

# Make predictions (shape: 3000,)
y_test_pred = stack.predict(X_test)

# Build the submission DataFrame
submission_ensemble = pd.DataFrame({
    "ID":    test["ID"].astype(int), 
    "price": np.expm1(y_test_pred)
})
submission_ensemble.to_csv("submission_ensemble_3.csv", index=False)
print(" Wrote submission_ensemble_3.csv (for Kaggle).")
```
