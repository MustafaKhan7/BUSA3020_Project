## 1.2: **Evaluation Criteria** 

The evaluation metric used in this competition is **Mean Absolute Error (MAE)**. MAE is a commonly used measure to evaluate the accuracy of regression-based forecasting models. It calculates the average of the absolute differences between predicted values and the actual values.

### Formula:

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

Where:  
- $n$ is the number of observations  
- $y_i$ is the actual value  
- $\hat{y}_i$ is the predicted value

**Pros -**
**Robust to outliers:**  
MAE treats all errors linearly, without giving disproportionate weight to large deviations. Unlike metrics such as Mean Squared Error (MSE), which square the residuals, MAE ensures that each unit of error contributes equally to the total. For example, if a prediction is off by 100, it contributes exactly 100 to the MAE. In contrast, under MSE, the same error would be squared to 10,000, heavily penalizing the model for a single large error. This can lead to a model being overly sensitive to outliers. MAE avoids this issue, making it more stable in datasets with extreme values.

**Cons -**
**Not scale-aware (no relative measure):**  
MAE does not consider the scale of the actual values. All errors are treated equally, regardless of the size of the true value. For example, an absolute error of \$300 is interpreted the same whether the actual price was \$500 or \$2,500. However, the relative impact is much larger in the first case (60%) compared to the second (12%). This limitation means that MAE may understate the severity of prediction errors for low-priced properties. In contrast, metrics like **Mean Absolute Percentage Error (MAPE)** incorporate relative error, making them more informative in contexts with large value ranges—such as rental markets with wide pricing variability.

## Task 1.4: Missing Values


```python
import pandas as pd

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

# count of missing values in each column of the training dataset
missing_train = train_df.isnull().sum()
missing_train = missing_train[missing_train > 0].sort_values(ascending=False)


print("Missing Values in train.csv:\n")
print(missing_train)

#count of missing values in each column of the test dataset
missing_test = test_df.isnull().sum()
missing_test = missing_test[missing_test > 0].sort_values(ascending=False)

print("Missing Values in test.csv:\n")
print(missing_test)
```

A considerable amount of missing data exists in both the training and test datasets. In the **training set**, high missingness is observed in features such as **`host_neighbourhood` (3,460)**, **`host_about` (2,270)**, **`neighbourhood`** and **`neighborhood_overview` (1,823 each)**, as well as **`host_location` (1,280)**. Relevant listing attributes such as **`bedrooms` (378)** and the **`review_scores_`** variables (with **22 missing values each**) also contain gaps that may impact model performance.

The **test set** shows similar patterns. Features like **`host_neighbourhood` (2,066)**, **`host_about` (1,441)**, **`neighbourhood`** and **`neighborhood_overview` (1,424 each)**, and **`host_location` (770)** are again among those with the most missing values. In addition, host-related metrics such as **`host_response_time`** and **`host_response_rate` (737 each)**, and **`host_acceptance_rate` (658)** have substantial missingness. Several review-related fields, including **`review_scores_*`**, as well as **`reviews_per_month`**, **`first_review`**, and **`last_review`**, each have around **263–300 missing entries**. The **`bedrooms`** field is also missing in **60** test records.

Most of the features with high missingness are **host-related or text-based**, which may not be essential for inclusion in a price prediction model. However, missing values in **`bedrooms`**, **`reviews_per_month`**, and **`review_scores_*`** are likely to influence pricing and should be addressed carefully during data preprocessing.


## Task 1.5: Univariate Data Characteristics

```python
train_df['price'] = train_df['price'].str.replace('$', '')
train_df['price'] = train_df['price'].str.replace(',', '')
```

```python
train_df['price'] = train_df['price'].astype(float)
print(train_df['price'].describe().round(2))
import matplotlib.pyplot as plt
```

```python
import numpy as np
import matplotlib.pyplot as plt

q1 = train_df['price'].quantile(0.25)
q3 = train_df['price'].quantile(0.75)
IQR = q3 - q1

# Defining Bounds
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR

print('Lower and Upper bound')
print(lower_bound, upper_bound)

#  prices under $500 
price_no_outliers = train_df[train_df['price'] < 500]['price']

bins = np.arange(25, 525, 50)
plt.hist(price_no_outliers, edgecolor='black', bins=bins)
plt.xticks([25, 75, 125, 175, 225, 275, 325, 375, 425, 475, 525])
plt.xlabel('Rental Cost Per Night ($)')
plt.ylabel('Count')
plt.title('Distribution of Airbnb Prices (Below $500)')
plt.show()
```

The `price` variable in the training dataset displays a wide range, with values ranging from **25** to **145,160**. The **mean price** is approximately **285.65**, but the distribution is notably right-skewed, as shown in the histogram. To improve readability, listings priced above **500** were excluded from the plot.

The **interquartile range (IQR)** spans from **117** (Q1) to **268** (Q3), meaning that 50% of listings fall within this price range. The histogram of filtered prices shows that most listings are concentrated between **100** and **200** per night, with frequency decreasing as the price increases. This indicates that the typical nightly rate for Airbnb listings in Melbourne falls within this moderate range.
