## Task 1.2: Evaluation Criteria â€“ Mean Absolute Error (MAE)

### Metric Used

The evaluation metric used in this competition is **Mean Absolute Error (MAE)**.  
MAE measures the average magnitude of the errors between predicted and actual values, without considering their direction.

### Advantages of MAE
- âœ… **Easy to interpret**: The error is in the same unit as the target variable (e.g., dollars).
- âœ… **Robust to outliers**: Unlike metrics that square the error, MAE treats all errors linearly and does not give extra weight to large deviations.
- âœ… **Intuitive for business stakeholders**: It directly reflects the average error in predictions.

### Limitations of MAE
- âŒ **Does not penalize large errors** as harshly as other metrics like MSE (Mean Squared Error), which may be important in some contexts.
- âŒ **Non-differentiable at zero**: While not a concern for this task, it can be less convenient for certain optimization algorithms in advanced model training.
