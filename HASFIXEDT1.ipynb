## 1.2: **Evaluation Criteria** 

The evaluation metric used in this competition is **Mean Absolute Error (MAE)**. MAE is a commonly used measure to evaluate the accuracy of regression-based forecasting models. It calculates the average of the absolute differences between predicted values and the actual values.

### Formula:

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

Where:  
- $n$ is the number of observations  
- $y_i$ is the actual value  
- $\hat{y}_i$ is the predicted value

**Pros -**
**Robust to outliers:**  
MAE treats all errors linearly, without giving disproportionate weight to large deviations. Unlike metrics such as Mean Squared Error (MSE), which square the residuals, MAE ensures that each unit of error contributes equally to the total. For example, if a prediction is off by 100, it contributes exactly 100 to the MAE. In contrast, under MSE, the same error would be squared to 10,000, heavily penalizing the model for a single large error. This can lead to a model being overly sensitive to outliers. MAE avoids this issue, making it more stable in datasets with extreme values.

**Cons -**
**Not scale-aware (no relative measure):**  
MAE does not consider the scale of the actual values. All errors are treated equally, regardless of the size of the true value. For example, an absolute error of 300 is interpreted the same whether the actual price was 500 or 2,500. However, the relative impact is much larger in the first case (60%) compared to the second (12%). This limitation means that MAE may understate the severity of prediction errors for low-priced properties. In contrast, metrics like **Mean Absolute Percentage Error (MAPE)** incorporate relative error, making them more informative in contexts with large value ranges—such as rental markets with wide pricing variability.
