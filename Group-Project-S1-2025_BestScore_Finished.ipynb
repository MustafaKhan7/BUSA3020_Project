## 1.2: **Evaluation Criteria** 

The evaluation metric used in this competition is **Mean Absolute Error (MAE)**. MAE is a commonly used measure to evaluate the accuracy of regression-based forecasting models. It calculates the average of the absolute differences between predicted values and the actual values.

### Formula:

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

Where:  
- $n$ is the number of observations  
- $y_i$ is the actual value  
- $\hat{y}_i$ is the predicted value

**Pros -**
**Robust to outliers:**  
MAE treats all errors linearly, without giving disproportionate weight to large deviations. Unlike metrics such as Mean Squared Error (MSE), which square the residuals, MAE ensures that each unit of error contributes equally to the total. For example, if a prediction is off by 100, it contributes exactly 100 to the MAE. In contrast, under MSE, the same error would be squared to 10,000, heavily penalizing the model for a single large error. This can lead to a model being overly sensitive to outliers. MAE avoids this issue, making it more stable in datasets with extreme values.

**Cons -**
**Not scale-aware (no relative measure):**  
MAE does not consider the scale of the actual values. All errors are treated equally, regardless of the size of the true value. For example, an absolute error of \$300 is interpreted the same whether the actual price was \$500 or \$2,500. However, the relative impact is much larger in the first case (60%) compared to the second (12%). This limitation means that MAE may understate the severity of prediction errors for low-priced properties. In contrast, metrics like **Mean Absolute Percentage Error (MAPE)** incorporate relative error, making them more informative in contexts with large value ranges—such as rental markets with wide pricing variability.

Finally, I applied one-hot encoding (dropping the first dummy) to all remaining nominal categorical columns so that they can be used directly in our model.

````python
## Task 2, Question 5 Code Here
from sklearn.preprocessing import StandardScaler


# drop colomns
train.drop(columns=['ID','name', 'description', 'neighborhood_overview','host_name', 'host_location','host_about','amenities','host_verifications', 'host_since', 'first_review','last_review','neighbourhood','source','host_neighbourhood'], inplace=True, errors='ignore')

test.drop(columns=[ 'name','description','neighborhood_overview','host_name','host_location','host_about','amenities','host_verifications','host_since','first_review','last_review','neighbourhood','source','host_neighbourhood'], inplace=True, errors='ignore')

#  Caping  outliers in price at 99th percentile in test dataset only since there is no price colomn in test dataset
p99_price = train['price'].quantile(0.99)
train['price'] = train['price'].clip(upper=p99_price)


#  capping  outliers in another feature at  99th percentile
p99_min_nights = train['minimum_nights'].quantile(0.99)
train['minimum_nights'] = train['minimum_nights'].clip(upper=p99_min_nights)
test['minimum_nights']  = test['minimum_nights'].clip(upper=p99_min_nights)

#log price
train['price'] = np.log1p(train['price'])

# Log transform skewed count 
train['number_of_reviews'] = np.log1p(train['number_of_reviews'])
test['number_of_reviews']  = np.log1p(test['number_of_reviews'])



# Standardise numeric features not including the target 'price'
numeric_feats = train.select_dtypes(include=['int64','float64']).columns.tolist()
numeric_feats.remove('price')  

scaler = StandardScaler()
train[numeric_feats] = scaler.fit_transform(train[numeric_feats])
test[numeric_feats]  = scaler.transform(test[numeric_feats])

````

Removed irrelevant text columns (like `name`, `description`, and `amenities`) that won’t be used in modeling, and dropped `ID` and other identifiers. Next, I capped extreme values for `price` and `minimum_nights` at their 99th percentiles to limit the influence of outliers. I then applied a log transform (`np.log1p`) to `price` and `number_of_reviews` to reduce skewness. Finally, I standardised all remaining numeric features (except the log-transformed `price`) using `StandardScaler` so that each has mean ≈ 0 and variance ≈ 1, ensuring the model treats them on the same scale.```

````python
#Task 3 Question 3

from sklearn.model_selection import KFold, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, RidgeCV, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, make_scorer
from sklearn.model_selection import cross_val_predict

#  Split the encoded train into X which are the features and y which is the target price

X = train.drop(columns=['price'])
y = train['price']


#   I set up a 5 fold cross validation and mae scorer
kf = KFold(n_splits=5, shuffle=True, random_state=42)
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)


# Simple linear regression
linreg = LinearRegression()

# fitting it once on the full train set to see coefficients
linreg.fit(X, y)

# extracting intercept and a few example coefficients
print("Simple LR intercept: ", round(linreg.intercept_, 2))

# combining feature names with their learned coefficients
lin_coefs = pd.Series(linreg.coef_, index=X.columns).sort_values(key=abs, ascending=False)

# top 10 linear Regression coefficients in multiplicative exp form
lin_exp_coefs = np.exp(lin_coefs.head(10))
print("\nTop 10 Linear Regression coefficient effects as multiplicative factor:")
print(lin_exp_coefs.round(3))


#  calucalting 5 fold CV MAE as a positive number (note to myself: 5 fold CV MAE is just the mean absolute error computed over five cross validation splits.it is same formual as MAE)
lin_preds = cross_val_predict(linreg, X, y, cv=kf, n_jobs=-1)
lin_mae = mean_absolute_error(np.expm1(y), np.expm1(lin_preds))
print(f"\nLinear Regression 5 fold CV MAE: {lin_mae:.2f}")


# ridge regression 
# using ridgeCV to try a grid of alpha values and pick the best by 5 fold Cross valifation
alphas = [0.01, 0.1, 1.0, 10.0, 100.0]

ridge_cv = RidgeCV(alphas=alphas, cv=kf, scoring=mae_scorer)
ridge_cv.fit(X, y)

best_alpha = ridge_cv.alpha_
print("\nRidge best alpha from cross validation:", best_alpha)

# refitting ridge  on the full training data using best alpha
ridge = Ridge(alpha=best_alpha)
ridge.fit(X, y)
ridge_coefs = pd.Series(ridge.coef_, index=X.columns).sort_values(key=abs, ascending=False)

ridge_exp_coefs = np.exp(ridge_coefs.head(10))
print("\nTop 10 Ridge coefficient effects as multiplicative factors:")
print(ridge_exp_coefs.round(3))

#  Computing 5 fold cross validation MAE for the  Ridge model
ridge = Ridge(alpha=best_alpha)
ridge.fit(X, y)
ridge_preds = cross_val_predict(ridge, X, y, cv=kf, n_jobs=-1)
ridge_mae = mean_absolute_error(np.expm1(y), np.expm1(ridge_preds))
print(f"\nRidge (α={best_alpha}) 5‐fold CV MAE: {ridge_mae:.2f}")


#  Random Forest Regression
rf = RandomForestRegressor(random_state=42)

param_grid = {
'n_estimators': [100, 200, 300],
'max_depth': [10, 15],
'min_samples_leaf':[1, 3]
}

grid_rf = GridSearchCV(
rf,
param_grid=param_grid,
scoring=mae_scorer,
cv=kf,
n_jobs=-1,
verbose=1
)
grid_rf.fit(X, y)

best_rf_params = grid_rf.best_params_
print("\nRandom Forest best parameters:", best_rf_params)
best_rf = grid_rf.best_estimator_


#  calculating 5 fold cross validation MAE of the best Random Forest
best_rf = grid_rf.best_estimator_
rf_preds = cross_val_predict(best_rf, X, y, cv=kf, n_jobs=-1)
rf_mae = mean_absolute_error(np.expm1(y), np.expm1(rf_preds))
print(f"\nRandom Forest 5 fold Cross Validation MAE: {rf_mae:.2f}")


#Just some checks to prepare for my regression model prediction

# making sure no object or texts remain
print(train.dtypes[train.dtypes == "object"])  
 
# double checking if price dropped from X
print(train.columns)     

 # scale of the target variable
print((train["price"].describe()))               
````

````python
#Task 3 question 4 ensemble method way 

#installing package for xgboost and lightgbm because i dont have it insatlled already
!pip install xgboost --quiet
!pip install lightgbm xgboost --quiet


from sklearn.model_selection import KFold, cross_val_score, GridSearchCV, cross_val_predict
from sklearn.linear_model import Ridge, ElasticNetCV, ElasticNet, LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from sklearn.metrics import mean_absolute_error, make_scorer
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor

# Splitting off X_full, y_full from train 

X_full = train.drop(columns=["price"])
y_full = train["price"].copy()

# The test dataFrame should still contain ID plus all dummy encoded columns, and i drop ID before predicting so that columns align exactly with X_full.columns.

X_test = test.drop(columns=["ID"], errors="ignore")

# Sanitising column names for XGBoost compatibility
X_full.columns = X_full.columns.str.replace(r'[\[\]<>]', '', regex=True)
X_test.columns = X_test.columns.str.replace(r'[\[\]<>]', '', regex=True)


# more checks 
print(" X_full shape:", X_full.shape)           
print(" Any missing in X_full? ", X_full.isna().sum().sum())  
print(" Any object columns in X_full? ", any(X_full.dtypes == "object"))



# Defining 5 fold cross validation and Mae scorer

kf = KFold(n_splits=5, shuffle=True, random_state=42)
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)


# Instantiating and fitting Ridge and Random Forest using best parameters form last question.I Write α =100 for Ridge and (200, 15, 3) for Random Forest parameters

# Ridge with α=100
ridge_best = Ridge(alpha=1.0, random_state=42)
ridge_best.fit(X_full, y_full)
ridge_preds = cross_val_predict(ridge_best, X_full, y_full, cv=kf, n_jobs=-1)
#Inverse Logging
ridge_mae = mean_absolute_error(np.expm1(y_full), np.expm1(ridge_preds))
print(f" Ridge 5-fold CV MAE: {ridge_mae:.2f}")


#  Random Forest with best tuned paramters( n estimators = 200, max depth =15, min samples leaf =1)
rf_best = RandomForestRegressor(
n_estimators=200,
max_depth=15,
min_samples_leaf=1,
random_state=42
)
rf_best.fit(X_full, y_full)
rf_preds = cross_val_predict(rf_best, X_full, y_full, cv=kf, n_jobs=-1)
#Inverse logging
rf_mae = mean_absolute_error(np.expm1(y_full), np.expm1(rf_preds))
print(f" RandomForest 5-fold CV MAE: {rf_mae:.2f}")


# Lgbm regression tuning ( finding best parameters)

param_grid_lgbm = {
"n_estimators": [100, 200, 300],
"learning_rate": [0.05, 0.1],
"max_depth": [3, 5],
"num_leaves": [15, 31]
}

grid_lgbm = GridSearchCV(
 LGBMRegressor(random_state=42),
 param_grid=param_grid_lgbm,
cv=kf,
 scoring=mae_scorer,
 n_jobs=-1
)

grid_lgbm.fit(X_full, y_full)
lgbm_best = grid_lgbm.best_estimator_
lgbm_preds = cross_val_predict(lgbm_best, X_full, y_full, cv=kf, n_jobs=-1)
lgbm_mae = mean_absolute_error(np.expm1(y_full), np.expm1(lgbm_preds))
print(f" LightGBM 5-fold CV MAE: {lgbm_mae:.2f}")


# gradient boost regression with gridsearchcv
gbr = GradientBoostingRegressor(random_state=42)

param_grid_gbr = {
"n_estimators": [100, 200, 300],
"learning_rate": [0.01, 0.05, 0.1],
"max_depth": [3, 5],
}
grid_gbr = GridSearchCV(
 gbr,
param_grid=param_grid_gbr,
cv=kf,
scoring=mae_scorer,
n_jobs=-1,
verbose=0
)
grid_gbr.fit(X_full, y_full)

best_gbr_params = grid_gbr.best_params_
print(f"\n GBR best paramater by cross validation: {best_gbr_params}")

gbr_best = grid_gbr.best_estimator_

# 5 fold Cross Validation MAE for GBR:
gbr_preds = cross_val_predict(gbr_best, X_full, y_full, cv=kf, n_jobs=-1)
gbr_mae = mean_absolute_error(np.expm1(y_full), np.expm1(gbr_preds))
print(f" GBR best Cross validation  MAE: {gbr_mae:.2f}")



#  Instantiating XGB and using squared error for objective
xgb = XGBRegressor(
random_state=42,
objective="reg:squarederror",  
verbosity=0
)

#  Defining a small grid to tune 
param_grid_xgb = {
"n_estimators": [100, 200, 300],
 "learning_rate": [0.03, 0.05, 0.1],
"max_depth": [3, 5],
"colsample_bytree": [0.8, 1.0]
}


grid_xgb = GridSearchCV(
xgb,
param_grid=param_grid_xgb,
cv=kf,
scoring=mae_scorer,
n_jobs=-1,
verbose=0
)
grid_xgb.fit(X_full, y_full)

best_xgb_params = grid_xgb.best_params_
print(f"\n XGBoost best paramaters by cross validation: {best_xgb_params}")

xgb_best = grid_xgb.best_estimator_
xgb_preds = cross_val_predict(xgb_best, X_full, y_full, cv=kf, n_jobs=-1)

#Inverse logging as denoted by np.expm1
xgb_mae = mean_absolute_error(np.expm1(y_full), np.expm1(xgb_preds))
print(f" XGBoost 5 fold CV MAE: {xgb_mae:.2f}")



#  Building the Stacked Model
# there is now four tuned base learners with best parameters, ridge_best, rf_best, gbr_best, lgbm_best, xgb_best,  i then stack them using ridgeCV as the final estimator.

stack_estimators = [
("ridge", ridge_best),
("rf",    rf_best),
("gbr",   gbr_best),
("lgbm",   lgbm_best),
("xgb",     xgb_best), 
]

stack = StackingRegressor(
 estimators       = stack_estimators,
 final_estimator  = RidgeCV(alphas=[0.01, 0.1, 1.0, 10.0]),
 cv               = KFold(n_splits=5, shuffle=True, random_state=42),
 n_jobs           = -1,
 passthrough      = False   
)


# 5 fold Cross Validation MAE for the stack

stack_preds = cross_val_predict(stack, X_full, y_full, cv=kf, n_jobs=-1)

#Inverse Logging
stack_mae = mean_absolute_error(np.expm1(y_full), np.expm1(stack_preds))
print(f"\n STACK (Ridge+RF+LGBM+GBR+XGB) 5-fold CV MAE: {stack_mae:.2f}")


#  refitting all models on the trainng dataset
ridge_best.fit(X_full, y_full)
rf_best.fit(X_full, y_full)
#enet_best.fit(X_full, y_full)
gbr_best.fit(X_full, y_full)
lgbm_best.fit(X_full, y_full)
xgb_best.fit(X_full, y_full)



# Re fitting the stacked ensemble on all the  7000 rows
stack.fit(X_full, y_full)



#  reindexing X_test to have exactly the same columns and order as X_full; this is very important
X_test = X_test.reindex(columns=X_full.columns, fill_value=0)

assert list(X_test.columns) == list(X_full.columns), ("‼ Columns must match EXACTLY before stack.predict!")

#  I predict on test set and save submission_ensemble csv file to folder
# This makes the predictions 
y_test_pred = stack.predict(X_test)

# Building the submission data frame
submission_ensemble = pd.DataFrame({
    "ID":    test["ID"].astype(int), 
    "price": np.expm1(y_test_pred)
})
submission_ensemble.to_csv("submission_ensemble_3.csv", index=False)
print(" Wrote submission_ensemble_3.csv (for Kaggle).")
````
