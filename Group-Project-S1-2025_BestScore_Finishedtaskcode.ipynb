## 1.2: **Evaluation Criteria** 

The evaluation metric used in this competition is **Mean Absolute Error (MAE)**. MAE is a commonly used measure to evaluate the accuracy of regression-based forecasting models. It calculates the average of the absolute differences between predicted values and the actual values.

### Formula:

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

Where:  
- $n$ is the number of observations  
- $y_i$ is the actual value  
- $\hat{y}_i$ is the predicted value

**Pros -**
**Robust to outliers:**  
MAE treats all errors linearly, without giving disproportionate weight to large deviations. Unlike metrics such as Mean Squared Error (MSE), which square the residuals, MAE ensures that each unit of error contributes equally to the total. For example, if a prediction is off by 100, it contributes exactly 100 to the MAE. In contrast, under MSE, the same error would be squared to 10,000, heavily penalizing the model for a single large error. This can lead to a model being overly sensitive to outliers. MAE avoids this issue, making it more stable in datasets with extreme values.

**Cons -**
**Not scale-aware (no relative measure):**  
MAE does not consider the scale of the actual values. All errors are treated equally, regardless of the size of the true value. For example, an absolute error of 300 is interpreted the same whether the actual price was 500 or 2,500. However, the relative impact is much larger in the first case (60%) compared to the second (12%). This limitation means that MAE may understate the severity of prediction errors for low-priced properties. In contrast, metrics like **Mean Absolute Percentage Error (MAPE)** incorporate relative error, making them more informative in contexts with large value ranges—such as rental markets with wide pricing variability.

**Task 2, Question 4**: Encode all categorical variables appropriately as discussed in class. 

- Where multiple values are given for an observation encode the observation as 'other'. 
- Where a categorical feature contains more than 5 unique values, map the features into 5 most frequent values + 'other' and then encode appropriately.  
(2 marks)

````python
## Task 2, Question 4 Code Here
#Finding 5 most freqent categories in training columns
import numpy as np

for col in cat_columns:
    # mark multi-valued cells as 'other'
    train.loc[train[col].str.contains(',', na=False), col] = 'other'
    test.loc[test[col].str.contains(',',  na=False), col] = 'other'

top5={}
for col in cat_columns:
      if train[col].nunique() > 5:
          top5[col] = train[col].value_counts().nlargest(5).index

for col in cat_columns:
    train[col] = train[col].astype(str)
    test[col]  = test[col].astype(str)

#Encoding Variables with more than 5 unique variables
for col in top5:
    train[col] = train[col].where(train[col].isin(top5[col]), 'Other')
    test[col] = test[col].where(test[col].isin(top5[col]), 'Other')

#Excluding host_response_time (ordinal variable) from cat_columns before one-hot encoding
cat_columns = list(cat_columns)
cat_columns = [col for col in cat_columns if col != 'host_response_time']

#Encoding Ordinal Variable (Host_response_time)
response_time_map = { 'within an hour': 1, 'within a few hours': 2, 'within a day': 3, 'a few days or more': 4}
train['host_response_time'] = train['host_response_time'].map(response_time_map)
test['host_response_time'] = test['host_response_time'].map(response_time_map)

#One hot encoding nominal categorical variables
train = pd.get_dummies(train, columns=cat_columns, drop_first=True)
test = pd.get_dummies(test, columns=cat_columns, drop_first=True)

````

In the question, I first encoded all variables with multiple values with the label 'other' as encoding these values would not be useful and likely negatively impact our model training, I then identifiyied the **top 5 most frequent levels** for each categorical column and converted all category columns to strings as a precaution to ensure appropriate encoding. For columns with **more than 5 unique values**, I grouped every level outside the top 5 into a single **“Other”** category. Next, I mapped the ordinal feature `host_response_time` into numeric values (1–4) according to increasing response duration and removed it from the list of columns to be one-hot encoded. Finally, I applied one-hot encoding (dropping the first dummy) to all remaining nominal categorical columns so that they can be used directly in our model.

## Task 3 Question 2






**Ordinary Least Squares Simple Linear Regression**

 
 A simple linear regression model assumes that the target (price) is a weighted sum of  features, plus a constant intercept where each β or coefficient is a number the model learns from the data


For e.g. predicted price=β0+β1⋅(bedrooms)+β2⋅(bathrooms)+…


This linear regression model is easy to understand in which there is one coefficient (β) for each feature or predictor and the easiest way to see how strongly each feature moves price up or down.  A disadvantage is it overfits if many correlated features are left in. Also, can’t model nonlinear effects for eg. if price goes up higher when there is a listing with at least 2 bedrooms in a certain neighbourhood, this simple linear regression model can not capture that nonlinear effect.
 


**Ridge Regression**


Ridge regression is similar to  simple linear regression but  it also adds a penalty that keeps all coefficients from getting too large. It is an easy to understand model and like the simple regression model (ordinary least squares), there is a coefficient for each feature.  And it is simple to modify to cross validate. The model also handles multicollinearity, for e.g. in  our dataset, features like bedrooms, beds, and accommodates are highly correlated. Without regularisation, models like the simple linear regression can assign vastly different coefficient values to each leading to an unstable model. Ridge shrinks those correlated coefficients toward one another creating better predictions on unseen data. One of disadvantages like simpler regression model is that it is still linear and ridge can’t find  non linear patterns.
 


**Random Forest Regression**


Random forest is a group of many decision trees. Each tree is built on a random sample of the data and with each split, it considers a random subset of features. For e.g. for the prediction of target variable price, features of Airbnb data are ran through every tree and then the individual tree predictions are averaged.
This is a nonlinear regression model so it can capture non linear relationships or patterns. For e.g. If price boosts radically once you have 3 plus bedrooms and are in a certain neighbourhood, a tree can split on  3 plus bedrooms first, then split on neighbourhood. The model will capture the pattern or effect  that only listings with 3+ bedrooms and in a  certain neighbourhood will  get that big price boost. In a linear model, it cannot capture that non linear effect for e.g in linear model, each extra bedroom  will always change the price by some fixed amount, regardless of neighbourhood.
Random forest can also the handle the many newly create dummy variables; it is also robust to outliers, so  a highly expensive Airbnb listing won’t skew the model’s predictions because each tree split is based on median values rather than the average in the subset data.


 **LightGBM**
 
 
 LightGBM, which stands for light gradient boosting machine, is a very fast and efficient version of a boosting algorithm. It builds trees one at a time, where each new tree focuses on fixing the mistakes of the previous ones. What makes lightGBM stand out is that it’s designed to handle large datasets and lots of categories quickly without using up too much memory. It can produce really accurate results with the right tuning, but it might overfit if you’re not careful, especially on smaller datasets.



**Gradient Boosting**


Gradient boosting is another tree based method that also builds trees one after another, with each tree improving on the last one by focusing more on the difficult to predict cases. It’s known for being more accurate than random forest when tuned properly, especially on more complex datasets. The downside is that it’s slower and more sensitive to how it’s configured, so you have to be more careful with tuning to avoid overfitting.



**XGBoost**


XGBoost, short for extreme gradient boosting, is a supercharged version of gradient boosting. It’s designed to be fast and accurate, with built in regularisation to help prevent overfitting and the ability to handle missing data. It’s widely used in competitions and by data scientists when they want to squeeze out the best performance. The trade off is that it can be a bit more complex to tune and understand, but it often gives top results when used well.



## Task 3 Question 3

````python
#Task 3 Question 3

from sklearn.model_selection import KFold, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, RidgeCV, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, make_scorer
from sklearn.model_selection import cross_val_predict

#  Split the encoded train into X which are the features and y which is the target price

X = train.drop(columns=['price'])
y = train['price']


#   I set up a 5 fold cross validation and mae scorer
kf = KFold(n_splits=5, shuffle=True, random_state=42)
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)


# Simple linear regression
linreg = LinearRegression()

# fitting it once on the full train set to see coefficients
linreg.fit(X, y)

# extracting intercept and a few example coefficients
print("Simple LR intercept: ", round(linreg.intercept_, 2))

# combining feature names with their learned coefficients
lin_coefs = pd.Series(linreg.coef_, index=X.columns).sort_values(key=abs, ascending=False)

# top 10 linear Regression coefficients in multiplicative exp form
lin_exp_coefs = np.exp(lin_coefs.head(10))
print("\nTop 10 Linear Regression coefficient effects as multiplicative factor:")
print(lin_exp_coefs.round(3))


#  calucalting 5 fold CV MAE as a positive number (note to myself: 5 fold CV MAE is just the mean absolute error computed over five cross validation splits.it is same formual as MAE)
lin_preds = cross_val_predict(linreg, X, y, cv=kf, n_jobs=-1)
lin_mae = mean_absolute_error(np.expm1(y), np.expm1(lin_preds))
print(f"\nLinear Regression 5 fold CV MAE: {lin_mae:.2f}")


# ridge regression 
# using ridgeCV to try a grid of alpha values and pick the best by 5 fold Cross valifation
alphas = [0.01, 0.1, 1.0, 10.0, 100.0]

ridge_cv = RidgeCV(alphas=alphas, cv=kf, scoring=mae_scorer)
ridge_cv.fit(X, y)

best_alpha = ridge_cv.alpha_
print("\nRidge best alpha from cross validation:", best_alpha)

# refitting ridge  on the full training data using best alpha
ridge = Ridge(alpha=best_alpha)
ridge.fit(X, y)
ridge_coefs = pd.Series(ridge.coef_, index=X.columns).sort_values(key=abs, ascending=False)

ridge_exp_coefs = np.exp(ridge_coefs.head(10))
print("\nTop 10 Ridge coefficient effects as multiplicative factors:")
print(ridge_exp_coefs.round(3))

#  Computing 5 fold cross validation MAE for the  Ridge model
ridge = Ridge(alpha=best_alpha)
ridge.fit(X, y)
ridge_preds = cross_val_predict(ridge, X, y, cv=kf, n_jobs=-1)
ridge_mae = mean_absolute_error(np.expm1(y), np.expm1(ridge_preds))
print(f"\nRidge (α={best_alpha}) 5‐fold CV MAE: {ridge_mae:.2f}")


#  random forest Regression
rf = RandomForestRegressor(random_state=42)

param_grid = {
'n_estimators': [100, 200, 300],
'max_depth': [10, 15],
'min_samples_leaf':[1, 3]
}

grid_rf = GridSearchCV(
rf,
param_grid=param_grid,
scoring=mae_scorer,
cv=kf,
n_jobs=-1,
verbose=1
)
grid_rf.fit(X, y)

best_rf_params = grid_rf.best_params_
print("\nRandom Forest best parameters:", best_rf_params)
best_rf = grid_rf.best_estimator_


#  calculating 5 fold cross validation MAE of the best Random Forest
best_rf = grid_rf.best_estimator_
rf_preds = cross_val_predict(best_rf, X, y, cv=kf, n_jobs=-1)
rf_mae = mean_absolute_error(np.expm1(y), np.expm1(rf_preds))
print(f"\nRandom Forest 5 fold Cross Validation MAE: {rf_mae:.2f}")


#just some checks to prepare for my regression model prediction

# making sure no object or texts remain
print(train.dtypes[train.dtypes == "object"])  
 
# double checking if price dropped from X
print(train.columns)     

 # scale of the target variable
print((train["price"].describe()))               
````
